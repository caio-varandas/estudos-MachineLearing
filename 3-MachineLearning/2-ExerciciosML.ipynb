```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

data, cluster = make_blobs(n_samples=400, n_features=2, centers=2, random_state=42)

for i in range(len(data)):

    if cluster[i] == 0:
        plt.scatter(data[i,0], data[i,1], marker='o', color="red")
    else:
        plt.scatter(data[i,0], data[i,1], marker='x', color="blue")


plt.xlabel('$X_1$')
plt.ylabel('$X_2$')

plt.show()
```
