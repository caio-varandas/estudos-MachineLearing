```python
#1 -

lista1 = []

for i in range(10):
    lista1.append(int(input(f'Digite o {i+1} valor da lista 1: ')))

lista2 = []

for i in range(10):
    lista2.append(int(input(f'Digite o {i+1} valor da lista 2: ')))

lista3 = []

for i in range(10):
    if i % 2 == 0:
        lista3.append(lista1[i])
    else:
        lista3.append(lista2[i])

print('Lista resultante:', lista3)
```

```python
#2 - 
import random

lista = []

for i in range(10):
    lista.append(random.randint(-100, 100))

#a)
media = sum(lista) / len(lista)
print('Média:', media)

#b)
maior = max(lista)
menor = min(lista)
print(f'Maior: {maior}, Menor: {menor}')

#c)
positivo = 0
negativo = 0
for i in lista:
    if i > 0:
        positivo += 1
    elif i < 0:
        negativo += 1

print(f'Positivos: {positivo}, Negativo: {negativo}')

#d)
abaixo = 0
acima = 0
for i in lista:
    if i > media:
        acima += 1
    elif i < media:
        abaixo += 1

print(f'Acima: {acima}, Abaixo: {abaixo}')

#e)
soma_quadrados = 0
for x in lista:
    soma_quadrados += (x - media) ** 2

desvio_padrao = (soma_quadrados / len(lista)) ** 0.5
print('Desvio padrão:', desvio_padrao)
```

```python
#6 - 

def media_notas(notas):
    return sum(notas) / len(notas)

entrada = input("Digite as notas separadas por espaço: ")
lista_notas = list(map(float, entrada.split()))
tupla_notas = tuple(lista_notas)

print(media_notas(tupla_notas))
```

```python
#8 - 

def conta_duplicados(lista):
    duplicados = 0
    for elem in lista:
        if lista.count(elem) > 1:
            duplicados += 1
    return duplicados


lista = [2, 2, 3, 3, 3]

tot = conta_duplicados(lista)

print(tot)
```
