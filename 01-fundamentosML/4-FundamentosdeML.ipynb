```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

#treino/teste

df = pd.read_csv(r"C:\Users\caiov\OneDrive\Desktop\house-price.txt", sep=",")

X = df.iloc[:, 0].values
y = df.iloc[:, -1].values

proporcoes = [0.1, 0.2, 0.25, 0.3, 0.4]

for i in proporcoes:

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=i,
        random_state=42
    )

    treino_gra = int((1-i) * 100)
    teste_gra = int(i * 100)

    plt.figure()

    plt.scatter(X_train, y_train, label="Treino")
    plt.scatter(X_test, y_test, label="Teste")
    plt.title(f"Divisão {treino_gra}/{teste_gra}")
    plt.xlabel("Feature")
    plt.ylabel("Preço")
    plt.grid()
    plt.legend()
    plt.show()

print(f"Tamanho do treino: {len(X_train)}")
print(f"Tamanho do teste: {len(X_test)}")

```
